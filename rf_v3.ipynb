```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
# reading training set
# this training set had duplicate
whole_training = pd.read_csv('training.csv')

#drop Category of the subject that's null
whole_training.dropna(inplace = True)

#msk = np.random.rand(len(whole_training)) < 0.8
#train = whole_training[msk]
#test = whole_training[~msk]

#data cleaning
whole_training["source"].replace('Google Play',0,inplace=True)
whole_training["source"].replace('App Store',1,inplace=True)
#whole_training.drop(['review_description', 'review_date',"roberta_comp'"], axis=1)
whole_training = whole_training[["source","rating","vader_neg","vader_neu","vader_pos","vader_compound", "roberta_neg","roberta_neu","roberta_pos","roberta_comp","category"]]
```

```python
whole_training
```

```python
#Splitting the data into independent and dependent variables
X = whole_training.iloc[:,0:10].values
y = whole_training.iloc[:,10].values
print('The independent features set: ')
print(X[:10,:])
print('The dependent variable: ')
print(y[:10])
```

```python
y
```

```python
# Feature Scaling
# use oob score, so no training data set
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
# Fitting Random Forest Classification to the Training set
classifier = RandomForestClassifier(oob_score=True, n_estimators = 10, criterion = 'entropy', random_state = 42)
classifier.fit(X, y)

print(classifier.oob_score_)
```

```python
# random forest
y_pred = classifier.predict(X)
#print(y_pred)
#print(y)
correct = (np.array(y_pred) == np.array(y))
accuracy = correct.sum() / correct.size
print(f"accuracy is {accuracy}\n")
print(pd.crosstab(y, y_pred, rownames=['Actual Category'], colnames=['Predicted Category']))
```
